```python
# import DiCE
import dice_ml
from dice_ml.utils import helpers # helper functions
```

```python
dataset = dice_ml.utils.helpers.load_adult_income_dataset()
```

```python
dataset.head()
```

```python
print(dataset)
```

```python
# description of transformed features
adult_info = dice_ml.utils.helpers.get_adult_data_info()
adult_info
```

```python
d = dice_ml.Data(dataframe=dataset, continuous_features=['age', 'hours_per_week'], outcome_name='income')
```

```python
ML_modelpath = dice_ml.utils.helpers.get_adult_income_modelpath()
m = dice_ml.Model(model_path= ML_modelpath)
```

```python
# initiate DiCE
exp = dice_ml.Dice(d, m)
```

```python
# query instance in the form of a dictionary; keys: feature name, values: feature value
query_instance = {'age':20, 
                  'workclass':'Private', 
                  'education':'School', 
                  'marital_status':'Single', 
                  'occupation':'Sales',
                  'race': 'White', 
                  'gender':'Male', 
                  'hours_per_week': 45}
```

```python
# generate counterfactuals
dice_exp = exp.generate_counterfactuals(query_instance, total_CFs=5, desired_class="opposite", features_to_vary=['age','education'])
```

```python
# visualize the results
dice_exp.visualize_as_dataframe()
```
